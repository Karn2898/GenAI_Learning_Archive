```python
from langchain import HugigngFacePipeline
from transformers import AutoTojenizer , pipeline
import torch

model="tiiuae/falcon-1.3b-instruct"

tokenizer=AutoTokenizer.from_pretrained(model)

pipeline=pipeline(
    "text-genetation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```
