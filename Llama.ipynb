```python
model_name="TheBloke/Llama-2-7B-chat-GGML"
model_basename="llama-2-7b-chat.ggmlv3.q5_1.bin"
```

# Import all the required libraries

```python
from huggingface_hub import hf_hub_download
```

```python
from llama_cpp import Llama
```

# download the model

if you are using small version of the model and not the quantized one then you can download it instead of using llama_cpp

```python
model_path=hf_hub_download(repo_id=model_name , filename=model_basename)
```

```python
model_path
```

Loading the model

```python
lcpp_llm=None
lcpp_llm=Llama(
    model_path=model_path,
    n_threads=2,
    n_batch=32,
    n_gpu_layers=32
)
```
